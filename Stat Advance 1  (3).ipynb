Q.2  What are the types of random variables? 
ANS: There are three principal types:

1. Discrete Random Variable
Takes on a countable number of distinct values (e.g. integers): like the number of heads in coin tosses or the number of emails received per day.
Probability described by a probability mass function (PMF), where the probabilities sum to 1.

3. Mixed Random Variable

Examples include waiting time at a traffic light (zero wait with some probability, otherwise positive continuous wait) or experiments combining a coin toss and spinner



 Q.3  What is the difference between discrete and continuous distributions? 
 ANS: Here‚Äôs how discrete and continuous probability distributions differ:

1. Nature of Outcomes
Discrete distributions model countable outcomes‚Äîeither finite or countably infinite. Examples include the roll of a die, number of children, or daily emails received.
Continuous distributions describe uncountably infinite outcomes, such as heights, weights, or exact measurement values on a continuous interval.

2. Probability Assignment
In a discrete distribution, each individual outcome x.
x has a nonzero probability, given by the probability mass function (PMF). The probabilities over all possible values sum to 1.
In a continuous distribution, the probability of the variable taking any specific exact value is zero. Instead, probability is assigned over intervals using the probability density function (PDF).

3. Mathematical Representation
Discrete: Probability is calculated via sums of PMF values. For instance, 
P(X=k)=f(k).
Individual point probabilities vanish.


4. Cumulative Distribution Function (CDF)
Discrete: The CDF is a step function, jumping at each possible value.
Continuous: The CDF is a smooth, continuous curve. If absolutely continuous, its derivative exists almost everywhere and equals the PDF.

5. Examples & Use Cases
Discrete:

Binomial (number of successes in independent trials)
Poisson (count of events in a time or space interval)


Continuous:

Normal distribution (e.g. heights, IQ scores)
Exponential, Uniform over intervals


6. Practical Implications
If you're modeling counts‚Äîsomething you can list or enumerate‚Äîuse a discrete distribution.
If you're modeling measurements‚Äîthings like time, length, or temperature‚Äîuse a continuous distribution.
Choose between discrete and continuous based on the type of data, not always how it's recorded‚Äîfor example, age in years (although discrete in recorded form) is often treated as continuous in analysis.

Q.11  What are the characteristics of a normal distribution?
ANS: Symmetry
The normal distribution is perfectly symmetric about its mean. This means the left and right halves of the distribution are mirror images of each other. Consequently, the mean, median, and mode of the distribution are all equal and located at the center.

Bell-Shaped Curve
The probability density function (PDF) of a normal distribution forms a bell-shaped curve, with the highest point at the mean. The curve tapers off symmetrically as it moves away from the mean, approaching but never touching the horizontal axis.

Empirical Rule (68-95-99.7 Rule)
For a normal distribution:
Approximately 68% of the data falls within one standard deviation of the mean.
Approximately 95% falls within two standard deviations.
Approximately 99.7% falls within three standard deviations.
This rule helps in understanding the spread and variability of data.

Asymptotic Nature
The tails of the normal distribution curve extend infinitely in both directions. Although they approach the horizontal axis, they never actually touch it, indicating that extreme values, while rare, are always possible.

Defined by Two Parameters
A normal distribution is fully characterized by its mean (Œº) and standard deviation (œÉ):
Mean (Œº): Determines the center of the distribution.
Standard Deviation (œÉ): Measures the spread or dispersion of the data.
These parameters dictate the shape and position of the distribution curve.

Skewness and Kurtosis
Skewness: The normal distribution has a skewness of 0, indicating no asymmetry.
Kurtosis: The normal distribution has a kurtosis of 3, which is considered mesokurtic, meaning it has a moderate peak and tails.

```python
Q.1 Write a Python program to generate a random variable and display its value.
ANS: import random

# Generate a random integer between 1 and 10 (inclusive)
random_integer = random.randint(1, 10)
print("Random Integer:", random_integer)

# Generate a random floating-point number between 0.0 and 1.0
random_float = random.random()
print("Random Float [0.0, 1.0):", random_float)

# Generate a random float in a specific range, e.g., between 5.0 and 15.0
random_uniform = random.uniform(5.0, 15.0)
print("Random Float [5.0, 15.0):", random_uniform)


Here‚Äôs what each function does:

random.randint(a, b) returns a random integer between a and b, inclusive 

random.random() generates a random floating‚Äëpoint number in the interval [0.0, 1.0) 

random.uniform(a, b) returns a random float between a and b, where a may be included and b is typically excluded  




Q.2   Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
ANS: import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the discrete uniform distribution parameters (like a fair six‚Äësided die)
a, b = 1, 6
dist = randint(low=a, high=b + 1)

# Define possible outcome values
values = np.arange(a, b + 1)

# Theoretical PMF values for each outcome
pmf_vals = dist.pmf(values)  # Equal probabilities of 1/(b-a+1)
# Empirical data: sample many random variates
samples = dist.rvs(size=1000)

# Plot both the empirical distribution (normalized histogram) and theoretical PMF
fig, ax = plt.subplots(figsize=(8, 5))
# Histogram approximates the PMF
ax.hist(samples, bins=np.arange(a, b + 2) - 0.5, density=True,
        alpha=0.6, color='gray', label='Empirical (histogram)')
# Theoretical PMF as points and vertical lines
ax.scatter(values, pmf_vals, color='red', zorder=5, label='Theoretical PMF')
ax.vlines(values, 0, pmf_vals, color='red', linestyles='--', zorder=4)

# Labels and appearance
ax.set_xticks(values)
ax.set_xlabel('Value')
ax.set_ylabel('Probability')
ax.set_title(f'Discrete Uniform Distribution PMF (from {a} to {b})')
ax.set_ylim(0, max(pmf_vals)*1.2)
ax.legend()
plt.show()






Q.3  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
ANS: def bernoulli_pmf(k, p):
    """
    Compute the probability mass function of a Bernoulli distribution.

    Parameters:
    - k: int (must be 0 or 1)
    - p: float in [0, 1], probability of success (i.e., P(X=1))

    Returns:
    - Probability P(X = k)
    """
    if k == 0:
        return 1 - p
    elif k == 1:
        return p
    else:
        return 0.0  # probability zero for any other k






Q.4 Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram
ANS: 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10
p = 0.5
size = 10000  # number of simulations

# Simulate binomial outcomes
data = np.random.binomial(n=n, p=p, size=size)

# Plot histogram (normalized to show probabilities)
plt.hist(data, bins=np.arange(-0.5, n+1.5, 1),
         density=True, edgecolor='black', alpha=0.7,
         label='Empirical histogram')

# Overlay theoretical PMF
x = np.arange(0, n+1)
pmf = binom.pmf(x, n=n, p=p)
plt.scatter(x, pmf, color='red', zorder=5, label='Theoretical PMF')
plt.vlines(x, 0, pmf, color='red', linestyle='dashed')

# Labels and styling
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(x)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()






Q.5   Create a Poisson distribution and visualize it using Python
ANS: import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lam = 4  # average rate Œª
size = 1000  # number of samples

# Simulate Poisson-distributed data
data = poisson.rvs(mu=lam, size=size)

# Plot histogram (normalized to show probabilities)
bins = np.arange(0, max(data) + 2) - 0.5
plt.hist(data, bins=bins, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Simulated data')

# Compute theoretical PMF
k = np.arange(0, max(data) + 1)
pmf = poisson.pmf(k, mu=lam)

# Overlay PMF
plt.vlines(k, 0, pmf, colors='red', linestyles='--', label='Theoretical PMF')
plt.scatter(k, pmf, color='red')

# Styling
plt.title(f'Poisson Distribution (Œª={lam})')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability')
plt.xticks(k)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()







Q.6 Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution
ANS:  import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the discrete uniform distribution parameters
a, b = 1, 6  # e.g., a fair six-sided die (1 through 6)
dist = randint(low=a, high=b+1)

# Create an array of possible outcome values
values = np.arange(a, b+1)

# Compute the cumulative distribution function (CDF) at each value
cdf_vals = dist.cdf(values)

# Plot the CDF as a step function
plt.figure(figsize=(8, 5))
plt.step(values, cdf_vals, where='post', label='Discrete Uniform CDF', color='blue', marker='o')
plt.title(f'CDF of Discrete Uniform Distribution [{a}, {b}]')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.xticks(values)
plt.ylim(0, 1.05)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()




Q.7 Generate a continuous uniform distribution using NumPy and visualize it
ANS:  import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Distribution parameters
a, b = 2.0, 5.0  # interval [a, b)
size = 10000     # number of random samples

# Generate samples
data = np.random.uniform(low=a, high=b, size=size)  # NumPy method :contentReference[oaicite:2]{index=2}

# Plot histogram of samples
plt.figure(figsize=(8, 5))
count, bins, _ = plt.hist(data, bins=50, density=True, color='skyblue',
                          edgecolor='black', alpha=0.7,
                          label='Sampled data')

# Overlay theoretical PDF
x = np.linspace(a, b, 200)
pdf = uniform.pdf(x, loc=a, scale=b - a)
plt.plot(x, pdf, 'r-', lw=2, label='Theoretical PDF')  # using scipy.stats.uniform :contentReference[oaicite:3]{index=3}

# Labels and aesthetics
plt.title(f'Continuous Uniform Distribution U({a}, {b})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()




Q.8 Simulate data from a normal distribution and plot its histogram
ANS: import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulation parameters
mean = 0.0         # Œº
std_dev = 1.0      # œÉ
size = 100_000     # number of samples

# Generate data
data = np.random.normal(loc=mean, scale=std_dev, size=size)
# Alternatively: data = norm.rvs(loc=mean, scale=std_dev, size=size)

# Plot histogram (normalized to form a PDF)
count, bins, _ = plt.hist(data, bins=50, density=True,
                          color='skyblue', alpha=0.6, edgecolor='black',
                          label='Simulated data')

# Compute theoretical PDF
x = np.linspace(bins.min(), bins.max(), 200)
pdf = norm.pdf(x, loc=mean, scale=std_dev)

# Overlay the PDF curve
plt.plot(x, pdf, color='red', lw=2, label='Theoretical PDF')

# Labels and styling
plt.title(f'Normal Distribution (Œº={mean}, œÉ={std_dev})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()




Q.9  Write a Python function to calculate Z-scores from a dataset and plot them
ANS:  import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def plot_z_scores(data, ddof=0):
    """
    Compute Z-scores for 'data' (1D array or list), plot histogram.

    Parameters:
      data : array-like
        Numeric data to standardize.
      ddof : int
        Degrees of freedom correction for standard deviation (default 0).
    """
    # Convert to NumPy array
    values = np.asarray(data)
    
    # Compute Z-scores using SciPy for robustness
    zs = zscore(values, ddof=ddof)  # mean and std internally handled :contentReference[oaicite:2]{index=2}

    # Plot the Z-score histogram
    plt.figure(figsize=(8, 5))
    plt.hist(zs, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title('Histogram of Z‚ÄëScores')
    plt.xlabel('Z‚ÄëScore')
    plt.ylabel('Density')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.show()
    
    return zs






Q.10   Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
ANS:  import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
np.random.seed(42)  # For reproducibility
population_size = 10000
sample_size = 30
num_samples = 1000

# Generate a population from an exponential distribution
population = np.random.exponential(scale=2.0, size=population_size)

# Collect sample means
sample_means = [np.mean(np.random.choice(population, size=sample_size)) for _ in range(num_samples)]

# Plot the population distribution
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(population, bins=50, density=True, alpha=0.6, color='g')
plt.title('Population Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Density')

# Plot the sampling distribution of the sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=50, density=True, alpha=0.6, color='b')
plt.title('Sampling Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Density')

# Overlay a normal distribution curve
mu, std = np.mean(sample_means), np.std(sample_means)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
plt.tight_layout()
plt.show()






Q.11 Simulate multiple samples from a normal distribution and verify the Central Limit Theorem
ANS: import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def simulate_clt(population, sample_sizes, num_samples=1000, seed=42):
    np.random.seed(seed)
    results = {}
    for n in sample_sizes:
        sample_means = [ np.mean(np.random.choice(population, size=n, replace=True))
            for _ in range(num_samples)]
        results[n] = np.array(sample_means)
    return results

def plot_clt_results(population, results):
    fig, axes = plt.subplots(1, len(results) + 1, figsize=(5 * (len(results) + 1), 4))
    # Plot population
    ax = axes[0]
    ax.hist(population, bins=50, density=True, alpha=0.6, color='gray')
    ax.set_title("Population Distribution")
    ax.set_xlabel("Value")
    ax.set_ylabel("Density")

    # Plot each sampling distribution
    for idx, (n, means) in enumerate(results.items(), start=1):
        ax = axes[idx]
        ax.hist(means, bins=30, density=True, alpha=0.6, label="Sample means (n={})".format(n))
        mu, sigma = means.mean(), means.std()
        x = np.linspace(mu - 4*sigma, mu + 4*sigma, 200)
        ax.plot(x, norm.pdf(x, mu, sigma), 'r--', lw=2, label="Normal fit")
        ax.set_title(f"Sampling dist (n={n})")
        ax.set_xlabel("Mean value")
        ax.set_ylabel("Density")
        ax.legend()
    plt.tight_layout()
    plt.show()

# Example usage:
# Generate a skewed non-normal population: exponential distribution
population = np.random.exponential(scale=2.0, size=10000)



Q.12 Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)
ANS: import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal():
    """
    Plot the standard normal distribution PDF: mean=0, std=1.
    """
    mu = 0
    sigma = 1
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
    y = norm.pdf(x, loc=mu, scale=sigma)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='Standard Normal PDF (Œº=0, œÉ=1)')
    plt.fill_between(x, y, alpha=0.2)
    plt.title('Standard Normal Distribution')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

    if __name__ == '__main__':
    plot_standard_normal()






Q.13 Generate random variables and calculate their corresponding probabilities using the binomial distribution
ANS: import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def simulate_and_compute_binomial(n, p, num_samples=1000):
    # Step 1: Generate binomial random variables (number of successes)
    samples = np.random.binomial(n=n, p=p, size=num_samples)
    
    # Step 2: Compute the theoretical PMF for all possible outcomes k = 0 to n
    k_values = np.arange(n + 1)
    pmf_values = binom.pmf(k_values, n, p)
    
    return samples, k_values, pmf_values

def plot_binomial_results(samples, k_values, pmf_values, n, p):
    plt.figure(figsize=(8, 5))
    
    # Empirical histogram of sample outcomes
    plt.hist(samples, bins=np.arange(n + 2) - 0.5, density=True,
             alpha=0.6, color='skyblue', edgecolor='black',
             label='Empirical frequencies')
    
    # Theoretical PMF overlay
    plt.scatter(k_values, pmf_values, color='red', zorder=5, label='Theoretical PMF')
    plt.vlines(k_values, 0, pmf_values, colors='red', linestyles='--', zorder=4)
    
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.xlabel('Number of Successes (k)')
    plt.ylabel('Probability')
    plt.xticks(k_values)
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Example usage:
n = 10        # number of trials
p = 0.3       # probability of success on each trial
num_samples = 10000






Q.14 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution
ANS: import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def z_score_probability(x, data, plot=True):
    
    Calculate Z-score for a data point and its corresponding standard normal probabilities.
    
    Parameters:
        x (float): data point
        data (array-like): list or array of observations (to compute mean and std)
        plot (bool): whether to plot the result
        
    Returns:
        z (float): the calculated Z-score
        p_left (float): P(Z_standard ‚â§ z)
        p_right (float): P(Z_standard ‚â• z)
    
    arr = np.asarray(data)
    mu = np.mean(arr)
    sigma = np.std(arr, ddof=0)  # population std by default
    z = (x - mu) / sigma
    
    # Standard normal cumulative probabilities
    p_left = norm.cdf(z)
    p_right = 1 - p_left
    
    if plot:
        xs = np.linspace(-4, 4, 1000)
        ys = norm.pdf(xs)
        plt.figure(figsize=(8, 5))
        plt.plot(xs, ys, label='Standard Normal PDF')
        # shade area below z
        mask = xs <= z
        plt.fill_between(xs, ys, where=mask, color='skyblue', alpha=0.5,
                         label=f"P(Z ‚â§ {z:.2f}) = {p_left:.3f}")
        plt.axvline(z, color='red', linestyle='--')
        plt.title(f"Standard Normal Distribution with Z = {z:.2f}")
        plt.xlabel("Z-value")
        plt.ylabel("Density")
        plt.legend()
        plt.grid(True)
        plt.show()
    
    return z, p_left, p_right






Q.15  Implement hypothesis testing using Z-statistics for a sample dataset
ANS: import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05, alternative='two-sided'):
    """
    Perform a one-sample Z-test.

    Parameters:
        sample (array-like): observed data
        pop_mean (float): mean under H0
        pop_std (float): known population standard deviation
        alpha (float): significance level (default 0.05)
        alternative (str): 'two-sided', 'greater', or 'less'

    Returns:
        z_stat (float): Z statistic
        p_value (float): p-value
    """
    sample = np.asarray(sample)
    n = len(sample)
    sample_mean = np.mean(sample)
    se = pop_std / np.sqrt(n)
    z_stat = (sample_mean - pop_mean) / se

    if alternative == 'two-sided':
        p_value = 2 * norm.sf(abs(z_stat))
        crit = norm.ppf(1 - alpha / 2)
    elif alternative == 'greater':
        p_value = norm.sf(z_stat)
        crit = norm.ppf(1 - alpha)
    else:  # 'less'
        p_value = norm.cdf(z_stat)
        crit = norm.ppf(alpha)

    return z_stat, p_value, crit

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=102, scale=10, size=40)  # sample
pop_mean = 100
pop_std = 15

z_stat, p_val, crit_value = z_test(sample_data, pop_mean, pop_std, alpha=0.05)
print(f"Z-statistic: {z_stat:.2f}, p-value: {p_val:.3f}, critical value: ¬±{crit_value:.2f}")





Q16 Create a confidence interval for a dataset using Python and interpret the result
ANS: import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def mean_confidence_interval(data, confidence=0.95):
    """
    Compute the confidence interval for the sample mean using t-distribution.
    Returns (mean, lower_bound, upper_bound).
    """
    arr = np.array(data)
    n = len(arr)
    mean = np.mean(arr)
    sem = stats.sem(arr)  # Standard error of the mean
    # Compute the critical t-value
    confidence_interval = stats.t.interval(alpha=confidence, df=n-1, loc=mean, scale=sem)
    return mean, confidence_interval[0], confidence_interval[1]

def plot_confidence_interval(data, confidence=0.95):
    mean, lower, upper = mean_confidence_interval(data, confidence)
    plt.figure(figsize=(8, 4))
    plt.errorbar(0, mean, yerr=[[mean - lower], [upper - mean]], fmt='o', capsize=10, color='blue')
    plt.xlim(-1, 1)
    plt.xticks([])
    plt.ylabel('Value')
    plt.title(f'{int(confidence*100)}% Confidence Interval for the Mean')
    plt.text(0, lower, f'{lower:.2f}', va='top', ha='center')
    plt.text(0, upper, f'{upper:.2f}', va='bottom', ha='center')
    plt.show()
    return mean, lower, upper

# Example data
data = [12, 15, 14, 10, 13, 17, 14, 15, 16, 14]
mean, lower, upper = plot_confidence_interval(data, confidence=0.95)
print(f"Sample Mean = {mean:.2f}")
print(f"95% Confidence Interval = ({lower:.2f}, {upper:.2f})")






Q.17 Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
ANS:  import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# 1. Generate random data from a normal distribution
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=50, scale=5, size=100)  # mean=50, std=5, n=100

# 2. Compute sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # unbiased estimator
n = len(data)

# 3. Calculate the critical z-value for 95% CI
confidence = 0.95
z_crit = stats.norm.ppf((1 + confidence) / 2)

# 4. Compute the margin of error
stderr = sample_std / np.sqrt(n)
margin = z_crit * stderr

# 5. Determine the confidence interval
ci_lower = sample_mean - margin
ci_upper = sample_mean + margin

# Display results
print(f"Sample mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")






Q.18 Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution
ANS: import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define parameters for the normal distribution
mu = 5       # mean
sigma = 2    # standard deviation

# Generate a range of x-values around the mean (¬±3œÉ)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 500)

# Compute the PDF values at each x
pdf_vals = norm.pdf(x, loc=mu, scale=sigma)

# Plot the PDF curve
plt.figure(figsize=(8, 5))
plt.plot(x, pdf_vals, label=f'N({mu}, {sigma}¬≤)')
plt.fill_between(x, pdf_vals, alpha=0.2)
plt.title('Normal Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()






Q.19  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution
ANS: import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_val = 7           # Average event rate (Œª)
k = np.arange(0, 17)     # Values from 0 to 16

# Compute Poisson CDF values
cdf_vals = poisson.cdf(k, mu=lambda_val)

# Display cumulative probabilities
for val, prob in zip(k, cdf_vals):
    print(f"P(X ‚â§ {val}) = {prob:.3f}")

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(k, cdf_vals, where='post', marker='o', color='purple')
plt.title(f'Poisson Distribution CDF (Œª = {lambda_val})')
plt.xlabel('k')
plt.ylabel('Cumulative Probability P(X ‚â§ k)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()






Q.20  Simulate a random variable using a continuous uniform distribution and calculate its expected value
ANS: import numpy as np

# Define interval
a, b = 2.0, 5.0

# Simulate random variables from Uniform(a, b)
np.random.seed(42)
samples = np.random.uniform(low=a, high=b, size=1000000)

# Empirical mean (simulation result)
empirical_mean = np.mean(samples)

# Theoretical mean
theoretical_mean = (a + b) / 2

print(f"Empirical Mean (simulation): {empirical_mean:.4f}")
print(f"Theoretical Expected Value:   {theoretical_mean:.4f}")







Q.21 Write a Python program to compare the standard deviations of two datasets and visualize the difference
ANS:  import numpy as np
import matplotlib.pyplot as plt

# Sample datasets
data_A = np.random.normal(loc=50, scale=5, size=1000)
data_B = np.random.normal(loc=50, scale=10, size=1000)

# Calculate standard deviations
std_A = np.std(data_A)
std_B = np.std(data_B)

# Print standard deviations
print(f"Standard Deviation of Data A: {std_A:.2f}")
print(f"Standard Deviation of Data B: {std_B:.2f}")

# Create histograms
plt.figure(figsize=(12, 6))

# Plot for Data A
plt.subplot(1, 2, 1)
plt.hist(data_A, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(np.mean(data_A), color='red', linestyle='dashed', linewidth=1)
plt.axvline(np.mean(data_A) + std_A, color='green', linestyle='dashed', linewidth=1)
plt.axvline(np.mean(data_A) - std_A, color='green', linestyle='dashed', linewidth=1)
plt.title(f"Data A: Mean & ¬±1 Std Dev")
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot for Data B
plt.subplot(1, 2, 2)
plt.hist(data_B, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
plt.axvline(np.mean(data_B), color='red', linestyle='dashed', linewidth=1)
plt.axvline(np.mean(data_B) + std_B, color='green', linestyle='dashed', linewidth=1)
plt.axvline(np.mean(data_B) - std_B, color='green', linestyle='dashed', linewidth=1)
plt.title(f"Data B: Mean & ¬±1 Std Dev")
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()







Q.22 Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
ANS: import numpy as np

# Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# Calculate Range
data_range = np.ptp(data)

# Calculate Interquartile Range (IQR)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print(f"Range: {data_range}")
print(f"Interquartile Range (IQR): {iqr}")





Q.23 Implement Z-score normalization on a dataset and visualize its transformation
ANS: import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Perform Z-score normalization
z_scores = (data - mean) / std_dev

# Plotting the original and normalized data
plt.figure(figsize=(12, 6))

# Original data histogram
plt.subplot(1, 2, 1)
sns.histplot(data, kde=True, color='blue', bins=30)
plt.title("Original Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Normalized data histogram
plt.subplot(1, 2, 2)
sns.histplot(z_scores, kde=True, color='green', bins=30)
plt.title("Z-Score Normalized Data")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()






Q.24  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.
ANS:import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    """
    Calculate the skewness and kurtosis of a dataset.

    Parameters:
    - data (array-like): The dataset to analyze.

    Returns:
    - tuple: Skewness and kurtosis values.
    """
    skewness_value = skew(data)
    kurtosis_value = kurtosis(data, fisher=True)  # Fisher's definition subtracts 3
    return skewness_value, kurtosis_value

# Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# Calculate skewness and kurtosis
skewness_value, kurtosis_value = calculate_skewness_kurtosis(data)

print(f"Skewness: {skewness_value:.4f}")
print(f"Kurtosis: {kurtosis_value:.4f}")
      




```
